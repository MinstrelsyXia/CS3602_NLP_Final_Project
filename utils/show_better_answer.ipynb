```python
import json
import os
def find_better_answer(test_dataset='ARC-e'):
    sft_json_path=rf".\version19_sft1\evals\sft\20241218_032709\predictions\checkpoint_14000_hf\{test_dataset}.json"
    plm_json_path=rf'.\version15_plm\evals\plm\20241216_070141\predictions\1_hf\{test_dataset}.json'
    with open(sft_json_path, 'r', encoding='utf-8') as sft_file:
                sft_json = json.load(sft_file)
    with open(plm_json_path, 'r', encoding='utf-8') as plm_file:
        plm_json = json.load(plm_file)
    question_idx={}  

    for key in sft_json:
        sft_ans = sft_json[key]
        plm_ans = plm_json[key]
        if (sft_ans["prediction"] == sft_ans["gold"]) and (plm_ans["prediction"] != plm_ans["gold"]):
            question_idx[key]= sft_ans["origin_prompt"]
            print(key,sft_ans["origin_prompt"])
    return question_idx
```
